```python
import mediapipe as mp
import cv2
import numpy as np
import os

mp_holistic = mp.solutions.holistic

folder_path = '/content/drive/MyDrive/DATASET_ug_sign_language/frames1'
output_folder = '/content/drive/MyDrive/DATASET_ug_sign_language/keypoints'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    for filename in os.listdir(folder_path):
        if filename.endswith(('.jpg', '.png')):
            filepath = os.path.join(folder_path, filename)
            image = cv2.imread(filepath)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            results = holistic.process(image)

            keypoints = []
            # Extract Pose Landmarks
            if results.pose_landmarks:
                keypoints.extend([[lmk.x, lmk.y, lmk.z] for lmk in results.pose_landmarks.landmark])
            else:
                keypoints.extend([[0, 0, 0]] * 33)  # Padding if not detected

            # Extract Hand Landmarks (Left & Right)
            for hand_landmarks in [results.left_hand_landmarks, results.right_hand_landmarks]:
                if hand_landmarks:
                    keypoints.extend([[lmk.x, lmk.y, lmk.z] for lmk in hand_landmarks.landmark])
                else:
                    keypoints.extend([[0, 0, 0]] * 21)  # Padding

            # Extract Face Landmarks (First 30 keypoints for simplification)
            if results.face_landmarks:
                keypoints.extend([[lmk.x, lmk.y, lmk.z] for lmk in results.face_landmarks.landmark[:30]])
            else:
                keypoints.extend([[0, 0, 0]] * 30)

            keypoints_array = np.array(keypoints).flatten()
            keypoints_array = np.nan_to_num(keypoints_array)

            output_filepath = os.path.join(output_folder, filename.split('.')[0] + ".npy")
            np.save(output_filepath, keypoints_array)
            print(f"Keypoints saved for {filename} to {output_filepath}")
```

```python
import os
import numpy as np
import gym
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from collections import deque
import random
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns  # For better confusion matrix visualization

# -------------------------------
#  Loading and Preprocessing Dataset
# -------------------------------
def load_dataset():
    dataset_path = "/content/drive/MyDrive/DATASET_ug_sign_language/keypoints"
    files = [np.load(os.path.join(dataset_path, file)) for file in os.listdir(dataset_path) if file.endswith('.npy')]
    max_length = max([len(f) for f in files])  # Get max length
    files = [np.pad(f, (0, max_length - len(f)), mode='constant') if f.ndim == 1 else
              np.pad(f, [(0, 0), (0, max_length - f.shape[1])], mode='constant') for f in files]
    return np.array(files)

dataset = load_dataset()
print(f"Loaded {len(dataset)} samples.")
print(f"First sample shape: {dataset[0].shape}")

# Normalize the keypoints
scaler = MinMaxScaler()
X = np.array([scaler.fit_transform(x.reshape(-1, x.shape[-1])).reshape(x.shape) for x in dataset])

# Splitting the data into training (80%) and testing (20%) sets
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")

# -------------------------------
#  Custom Gym Environment for DQN (using training data)
# -------------------------------
class SignLanguageEnv(gym.Env):
    def __init__(self):
        super(SignLanguageEnv, self).__init__()
        self.observation_space = gym.spaces.Box(low=0, high=1, shape=X_train[0].shape, dtype=np.float32)
        self.num_classes = 10  # Initial Number of classes from your dataset labels
        self.action_space = gym.spaces.Discrete(self.num_classes)
        self.data = X_train
        self.current_index = 0

    def reset(self):
        self.current_index = 0
        return self.data[self.current_index]

    def step(self, action):
        self.current_index += 1
        done = self.current_index >= len(self.data) - 1
        # Correct action is the class with the maximum sum of keypoints
        correct_action = np.argmax(self.data[self.current_index].sum(axis=(0))) % self.num_classes
        reward = 2 if action == correct_action else -0.1
        if done:
            self.reset()
        return self.data[self.current_index], reward, done, {}

env = SignLanguageEnv()
print("Custom Gym environment initialized!")

# -------------------------------
#  Build DQN Model
# -------------------------------
def build_model(input_shape, num_actions):
    model = Sequential([
        tf.keras.layers.Input(shape=input_shape),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(num_actions, activation='linear')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse')
    return model

# -------------------------------
#  Experience Replay Buffer
# -------------------------------
class ReplayBuffer:
    def __init__(self, capacity=5000):
        self.buffer = deque(maxlen=capacity)
    def add(self, experience):
        self.buffer.append(experience)
    def sample(self, batch_size):
        return random.sample(self.buffer, batch_size)
    def size(self):
        return len(self.buffer)

# -------------------------------
#  Train DQN Agent
# -------------------------------
def train_dqn(env, episodes=25, batch_size=16, gamma=0.99, epsilon=1.0, epsilon_decay=0.98, min_epsilon=0.1):
    input_shape = env.observation_space.shape
    num_actions = env.action_space.n
    model = build_model(input_shape, num_actions)
    target_model = build_model(input_shape, num_actions)
    target_model.set_weights(model.get_weights())
    replay_buffer = ReplayBuffer()
    total_rewards_per_episode = []  # Store total reward for each episode

    for episode in range(episodes):
        state = env.reset()
        total_reward = 0
        done = False
        step_count = 0
        max_steps = 50

        while not done and step_count < max_steps:
            if np.random.rand() < epsilon:
                action = env.action_space.sample()
            else:
                q_values = model.predict(np.expand_dims(state, axis=0), verbose=0)
                action = np.argmax(q_values[0])
            next_state, reward, done, _ = env.step(action)
            replay_buffer.add((state, action, reward, next_state, done))
            state = next_state
            total_reward += reward
            step_count += 1

            if replay_buffer.size() >= batch_size:
                batch = replay_buffer.sample(batch_size)
                states, actions, rewards, next_states, dones = zip(*batch)
                states = np.array(states)
                next_states = np.array(next_states)
                targets = model.predict(states, verbose=0)
                next_q_values = target_model.predict(next_states, verbose=0)
                for i in range(batch_size):
                    targets[i, actions[i]] = rewards[i] + gamma * np.max(next_q_values[i]) * (1 - dones[i])
                model.train_on_batch(states, targets)
        if episode % 5 == 0:
            target_model.set_weights(model.get_weights())
        epsilon = max(min_epsilon, epsilon * epsilon_decay)
        print(f"Episode {episode + 1}, Total Reward: {total_reward}, Epsilon: {epsilon:.4f}")
        total_rewards_per_episode.append(total_reward)

    # Plotting the rewards
    plt.plot(total_rewards_per_episode)
    plt.xlabel("Episode")
    plt.ylabel("Total Reward")
    plt.title("Total Reward per Episode during Training")
    plt.show()

    return model, total_rewards_per_episode

# -------------------------------
#  Evaluation Functions (Corrected)
# -------------------------------
def evaluate_model(model, env, episodes=5):
    correct_predictions = 0
    total_predictions = 0
    all_true = []  # Collect true labels
    all_preds = []  # Collect predicted labels

    for episode in range(episodes):
        state = env.reset()
        done = False
        while not done:
            q_values = model.predict(np.expand_dims(state, axis=0), verbose=0)
            action = np.argmax(q_values[0])
            # The ground truth label should be derived in the same way as in env.step
            correct_action = np.argmax(state.sum(axis=(0))) % env.num_classes

            all_true.append(correct_action)
            all_preds.append(action)

            if action == correct_action:
                correct_predictions += 1
            total_predictions += 1
            state, _, done, _ = env.step(action)

    accuracy = accuracy_score(all_true, all_preds)  # Use sklearn's accuracy_score
    print(f"Model Accuracy: {accuracy * 100:.2f}%")

def evaluate_model_extended(model, env, episodes=5):
    all_true = []
    all_preds = []
    total_reward = 0

    for episode in range(episodes):
        state = env.reset()
        done = False
        while not done:
            q_values = model.predict(np.expand_dims(state, axis=0), verbose=0)
            pred = np.argmax(q_values[0])
            true_label = np.argmax(state.sum(axis=(0))) % env.num_classes

            all_preds.append(pred)
            all_true.append(true_label)
            state, reward, done, _ = env.step(pred)
            total_reward += reward

    # Determine actual number of classes
    num_classes_actual = len(np.unique(np.concatenate((all_true, all_preds))))

    conf_mat = confusion_matrix(all_true, all_preds, labels=range(num_classes_actual))  # Ensure correct size
    precision = precision_score(all_true, all_preds, average='macro', zero_division=0)

    specificity_list = []
    for i in range(num_classes_actual):
        TP = conf_mat[i, i]
        FP = conf_mat[:, i].sum() - TP
        FN = conf_mat[i, :].sum() - TP
        TN = conf_mat.sum() - (TP + FP + FN)
        specificity_i = TN / (TN + FP) if (TN + FP) > 0 else 0
        specificity_list.append(specificity_i)
    avg_specificity = np.mean(specificity_list)

    print("Confusion Matrix:")
    print(conf_mat)

    # Visualize the Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_mat, annot=True, fmt="d", cmap="Blues")  # Use seaborn for better visualization
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix")
    plt.show()

    print(f"Precision (macro-average): {precision:.2f}")
    print(f"Average Specificity: {avg_specificity:.2f}")
    print(f"Total Reward over evaluation: {total_reward}")
    return conf_mat, precision, avg_specificity

# -------------------------------
#  Train, Save, Load, and Evaluate the Model
# -------------------------------

# Train the model
trained_model, training_rewards = train_dqn(env)

# Save the trained model using the native Keras format
trained_model.save("trained_dqn_model.keras")
print("Model saved to trained_dqn_model.keras")

# Load the saved model from disk
loaded_model = tf.keras.models.load_model("trained_dqn_model.keras")
print("Model loaded successfully from disk for evaluation.")

# Evaluate the loaded model using both simple accuracy and extended metrics
print("\nSimple Evaluation:")
evaluate_model(loaded_model, env, episodes=5)

print("\nExtended Evaluation:")
evaluate_model_extended(loaded_model, env, episodes=5)
```
